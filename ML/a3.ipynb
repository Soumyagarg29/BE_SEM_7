```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Advait0801/SPPU-COMP-2019-Assignments/refs/heads/ML/Neural%20Network/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
gender = pd.get_dummies(df['Gender'] , drop_first=True)
geograpy = pd.get_dummies(df['Geography'] , drop_first=True)
df = pd.concat([df , gender , geograpy] , axis=1)
df =df.drop(['Gender' , 'Geography'] ,axis=1)
df = df.dropna()
df
```

```python
X = df.drop('Exited' , axis = 1) # axis = 1 is to drop te entire column
Y=df['Exited']

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X,Y)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(activation='relu' , units = 6))#input layer 
model.add(Dense(activation='relu' , units = 6)) #hidden layer
model.add(Dense(activation='sigmoid', units = 1))  #for classification
model.compile(optimizer='adam' , loss='binary_crossentropy')
model.fit(x_train , y_train , batch_size = 10 , epochs = 50)
```

```python
y_pred = model.predict(x_test)
y_pred = y_pred > 0.5
```
