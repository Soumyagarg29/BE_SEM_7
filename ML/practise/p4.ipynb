```python
learning_rate = 0.01
step_size =1
precision = 0.001
initial_x = 2
df = lambda x : 2 * (x+3)  
iterations = 0
max_iterations = 10000
gradient_descent = []
x_history = []
f = lambda x : (x+3)**2

```

```python
while step_size > precision and iterations < max_iterations :
    prev = initial_x
    initial_x-= learning_rate * df(prev)
    step_size = abs(prev - initial_x)
    x_history.append(prev)
    gradient_descent.append(initial_x)
    iterations+=1

```

```python
print("Local Minima Occur at : " , initial_x)
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.scatterplot(gradient_descent)
```

```python
def function(x):
    return (x+3)**2

def plotCurve():
    x_values = np.linspace(-10, 5, 100)
    y_values = function(x_values)

    # Plot the function and gradient descent steps
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label='y = (x + 3)^2', color='blue')

def plotGradientPoints() :
    sns.scatterplot(x=x_history, y=[df(x) for x in x_history], color='red', marker='o', s=100, label='Gradient Descent Steps')


plotCurve()
plotGradientPoints()
```
